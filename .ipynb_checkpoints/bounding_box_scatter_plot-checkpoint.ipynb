```python
import xml.etree.ElementTree as ET
import seaborn as sns
import matplotlib.pyplot as plt

def read_annotation(annotation_path):
    tree = ET.parse(annotation_path)
    root = tree.getroot()

    boxes = []
    labels = []
    for obj in root.findall('object'):
        label = obj.find('name').text
        labels.append(label)
        bbox = obj.find('bndbox')
        xmin = int(bbox.find('xmin').text)
        ymin = int(bbox.find('ymin').text)
        xmax = int(bbox.find('xmax').text)
        ymax = int(bbox.find('ymax').text)
        boxes.append([(xmin + xmax) / 2, (ymin + ymax) / 2])  # Use center coordinates

    return boxes, labels

def visualize_scatter_plot(boxes, labels):
    x, y = zip(*boxes)

    # Get unique class labels and assign a unique color to each class
    unique_labels = list(set(labels))
    num_classes = len(unique_labels)
    class_colors = sns.color_palette("tab20", num_classes)

    # Map class labels to their corresponding colors
    label_to_color = {label: color for label, color in zip(unique_labels, class_colors)}

    # Convert class labels to colors for each data point
    colors = [label_to_color[label] for label in labels]

    # Set up the seaborn style
    sns.set(style="whitegrid", font_scale=1.2)

    # Create a scatter plot
    plt.figure(figsize=(10, 8))
    ax = sns.scatterplot(x=x, y=y, hue=labels, palette=label_to_color, alpha=0.8, s=100, edgecolor='none')

    # Customizations
    plt.xlabel('X', fontweight='bold', fontsize=14, labelpad=10)
    plt.ylabel('Y', fontweight='bold', fontsize=14, labelpad=10)
    plt.title('Pascal VOC Dataset Scatter Plot', fontweight='bold', fontsize=18, pad=20)
    plt.legend(title='Class', title_fontsize=12, loc='upper left', fontsize=12, markerscale=1.5)

    # Remove spines
    sns.despine()

    # Set grid lines and background color
    plt.grid(True, linewidth=0.5, linestyle='--', alpha=0.5)
    ax.set_facecolor('whitesmoke')

    # Tight layout
    plt.tight_layout()

    # Remove legend border
    legend = ax.get_legend()
    legend.legendPatch.set_edgecolor('none')

    plt.show()
    
# Function to read annotations from multiple files in a folder
def read_annotations_from_folder(folder_path, num_annotations=10):
    boxes_list = []
    labels_list = []
    for i in range(1, num_annotations + 1):
        annotation_file = f"image{i}.xml"
        annotation_path = os.path.join(folder_path, annotation_file)
        if os.path.exists(annotation_path):
            boxes, labels = read_annotation(annotation_path)
            if boxes:  # Check if the boxes list is not empty
                boxes_list.extend(boxes)
                labels_list.extend(labels)
    return boxes_list, labels_list

# Usage example
folder_path = 'pascal_voc_datasets/VOCdevkit/PlotsEnchanced_Original_NoAugmentation/Annotations'
boxes, labels = read_annotations_from_folder(folder_path, num_annotations=10000)

# Check if any annotations were found before plotting
if boxes and labels:
    visualize_scatter_plot(boxes, labels)
else:
    print("No annotations found.")
```

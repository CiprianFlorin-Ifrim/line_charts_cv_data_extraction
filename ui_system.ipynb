```python
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk
import cv2
import numpy as np
from mss import mss
import pyautogui
import threading
import time

# Global variable to control the recording thread
recording = False

# Functions
def open_options():
    root.withdraw()
    options_page.deiconify()

def show_description():
    messagebox.showinfo("System Description", "This is the system description.")

def quit_program():
    root.destroy()

def choose_image():
    filename = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg *.png *.bmp *.jpeg *.tiff")])
    # do something with filename

def choose_video():
    filename = filedialog.askopenfilename(filetypes=[("Video files", "*.mp4 *.avi *.mov *.wmv")])
    # do something with filename

def desktop_capture():
    global recording
    recording = True
    threading.Thread(target=record_screen).start()

def stop_capture():
    global recording
    recording = False

def record_screen():
    screen_size = (1920, 1080)  # Set this to your screen resolution
    fourcc = cv2.VideoWriter_fourcc(*"XVID")
    out = cv2.VideoWriter("screen_capture.avi", fourcc, 20.0, screen_size)
    sct = mss()

    global recording
    while recording:
        img = sct.grab({"top": 0, "left": 0, "width": screen_size[0], "height": screen_size[1]})
        frame = np.array(img)
        frame = cv2.cvtColor(frame, cv2.COLOR_RGBA2BGR)
        
        # Apply some operation to the frame here.
        # For instance, let's convert it to grayscale.
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # If you want to view the processed frame in real-time, uncomment the following line:
        cv2.imshow('Screen Capture', gray)

        # And if you press 'q' on your keyboard, the loop will break
        if cv2.waitKey(1) == ord('q'):
            break

        out.write(frame)
    
    # Release the VideoWriter and close OpenCV windows
    out.release()
    cv2.destroyAllWindows()

def back_to_main():
    options_page.withdraw()
    root.deiconify()

# Create main page
root = tk.Tk()
root.title('Main Page')

img = ImageTk.PhotoImage(Image.open("ui_backgrounds/page_1_background.jpg"))  # Use your own image here
background_label = tk.Label(root, image=img)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

btn1 = tk.Button(root, text="System options", command=open_options)
btn1.pack(padx=20, pady=20)

btn2 = tk.Button(root, text="System description", command=show_description)
btn2.pack(padx=20, pady=20)

btn3 = tk.Button(root, text="Quit program", command=quit_program)
btn3.pack(padx=20, pady=20)

# Create options page
options_page = tk.Toplevel(root)
options_page.title('Options Page')
options_page.withdraw()

img2 = ImageTk.PhotoImage(Image.open("ui_backgrounds/page_1_background.jpg"))  # Use your own image here
background_label2 = tk.Label(options_page, image=img2)
background_label2.place(x=0, y=0, relwidth=1, relheight=1)

opt_btn1 = tk.Button(options_page, text="Choose image", command=choose_image)
opt_btn1.pack(padx=20, pady=20)

opt_btn2 = tk.Button(options_page, text="Choose video", command=choose_video)
opt_btn2.pack(padx=20, pady=20)

opt_btn3 = tk.Button(options_page, text="Desktop capture", command=desktop_capture)
opt_btn3.pack(padx=20, pady=20)

opt_btn_stop = tk.Button(options_page, text="Stop Capture", command=stop_capture)
opt_btn_stop.pack(padx=20, pady=20)

opt_btn4 = tk.Button(options_page, text="Back", command=back_to_main)
opt_btn4.pack(padx=20, pady=20)

root.mainloop()
```

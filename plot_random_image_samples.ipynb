```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def display_random_images(folder_path, save_path, grid_size=(3, 3)):
    # Get a list of all image files in the specified folder
    image_files = [f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

    # Select random images
    random_images = random.sample(image_files, min(grid_size[0] * grid_size[1], len(image_files)))

    # Create a new figure
    fig, axes = plt.subplots(grid_size[0], grid_size[1], figsize=(20, 20))

    for i, ax in enumerate(axes.flat):
        if i < len(random_images):
            img_path = os.path.join(folder_path, random_images[i])
            img = mpimg.imread(img_path)
            ax.imshow(img)
            filename_without_extension = os.path.splitext(random_images[i])[0]
            ax.set_xlabel(filename_without_extension)
        else:
            ax.axis('off') # Hide axes for empty subplots

    plt.tight_layout()
    
    # Save the figure as a PNG image at 300 DPI
    output_file = os.path.join(save_path, 'random_images.png')
    plt.savefig(output_file, dpi=300)
    
    plt.show()
    print(f"Image saved at {output_file}")


folder_path = 'pascal_voc_datasets/VOCdevkit/PlotsEnchanced_Original_NoAugmentation/JPEGImages' 
save_path = 'findings/plotted_images'   
display_random_images(folder_path, save_path)
```

```python
import os
from PIL import Image
from tqdm import tqdm
from joblib import Parallel, delayed

src_dir = 'pascal_voc_datasets/VOCdevkit/PlotsEnchanced_Original_NoAugmentation/JPEGImages'
dst_dir = 'pascal_voc_datasets/VOCdevkit/PlotsEnchanced_Downsampled_NoAugmentation/JPEGImages'

os.makedirs(dst_dir, exist_ok=True)

def resize_image(filename):
    img = Image.open(os.path.join(src_dir, filename))
    width, height = img.size
    if width < height:
        new_width = 512
        new_height = int(new_width * height / width)
    else:
        new_height = 512
        new_width = int(new_height * width / height)
    
    img_resized = img.resize((new_width, new_height), Image.LANCZOS)
    img_resized.save(os.path.join(dst_dir, filename))

image_files = [filename for filename in os.listdir(src_dir) if filename.endswith('.jpg')]

# Using all available cores
Parallel(n_jobs=-1)(
    delayed(resize_image)(filename) for filename in tqdm(image_files, desc="Resizing images")
)
```

```python
from PIL import Image, ImageDraw, ImageFont, ImageSequence

# Load the GIF file
input_gif_path = "ui/processing/original.gif" 
output_gif_path = "ui/processing/output.gif"
text = "Progress"
font_path = "impact.ttf" # Replace with the actual path to the Impact font file
font_size = 32
font_color = (255, 255, 255)  # White color
outline_color = (0, 0, 0)      # Black outline color
text_distance_from_edge = 10   # 10 pixels from the edge
frames_per_text = 4

def add_text_to_frame(frame, text):
    draw = ImageDraw.Draw(frame)
    font = ImageFont.truetype(font_path, font_size)
    text_width, text_height = draw.textsize(text, font=font)
    frame_width, frame_height = frame.size
    text_position = ((frame_width - text_width) // 2, frame_height - text_distance_from_edge - text_height)
    
    # Draw black outline
    for x_offset in [-1, 0, 1]:
        for y_offset in [-1, 0, 1]:
            draw.text((text_position[0] + x_offset, text_position[1] + y_offset), text, font=font, fill=outline_color)
    
    # Draw white text
    draw.text(text_position, text, font=font, fill=font_color)

# Open the input GIF
with Image.open(input_gif_path) as img:
    frames = []
    for frame in ImageSequence.Iterator(img):
        frame_copy = frame.copy()
        
        # Get the original frame duration
        duration = frame.info.get('duration', 100)  # Default to 100ms
        
        # Add text to each frame
        add_text_to_frame(frame_copy, text)
        frames.extend([frame_copy] * frames_per_text)
        
        # Set the duration for each modified frame
        for _ in range(frames_per_text):
            frame_copy.info['duration'] = duration
        
    # Save the modified frames as a new GIF
    frames[0].save(output_gif_path, save_all=True, append_images=frames[1:], loop=0, duration=duration)
    
print("GIF processing complete.")
```

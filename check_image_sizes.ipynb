```python
import os
from PIL import Image
from collections import defaultdict

#-------------------------------------------------------------------------------------JUPYTER NOTEBOOK SETTINGS-------------------------------------------------------------------------------------
from IPython.core.display import display, HTML                                    
display(HTML("<style>.container { width:100% !important; }</style>"))     
```

```python
# Absolute path to your image directory
#image_dir = '/absolute/path/to/your/images'  # Unix-based systems
image_dir = 'D:\\Academia\\webpage_plot_scraping_backup\\sample_plots\\coindesk.com'  # Windows

# Get a list of all image file paths
image_files = [os.path.join(image_dir, file) for file in os.listdir(image_dir) if file.endswith(('.png', '.jpg', '.jpeg'))]

# Initialize width and height variables
width, height = None, None

diff_size_counter = 0
# Go through all image files
for image_file in image_files:
    # Open the image file
    with Image.open(image_file) as img:
        # If width and height are not yet set, set them
        if width is None and height is None:
            width, height = img.size
        # If width and height don't match the size of the current image, print a warning
        elif (width, height) != img.size:
            print(f'Image {image_file} has a different size: {img.size} instead of {(width, height)}')
            diff_size_counter += 1

print('Checked all images, the number of images with a different size is: ', diff_size_counter)
```
